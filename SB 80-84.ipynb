```python
p = [0.2, 0.3, 0.5]
n = [3, 5, 10]
promedio_retornos = calculate_weighted_returns(SB, p, n)

# Encuentra la fila con el valor más alto en 'Weighted Average Return'
fila_superior = promedio_retornos.sort_values(by='Weighted Average Return', ascending=False).iloc[0]

# Extrae el nombre y el valor
nombre = fila_superior['Siefore']
valor_alto = fila_superior['Weighted Average Return']

print(f"El siefore: {nombre}, tiene el rendimiento más alto: {valor_alto}")
promedio_retornos
```
