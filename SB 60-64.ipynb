```python
SB, SB_inv = load_and_process_excel('SB 60-64','WEB_SB 60-64')
```

Ahora calcularemos el promedio del retorno segun Cálculo conforme a las DISPOSICIONES de Carácter General  que establecen el procedimiento para la construcción de los indicadores de rendimiento neto de las sociedades de inversión especializadas de fondos para el retiro publicadas en el DOF el 2 de junio de 2022 y sus posteriores modificaciones.

Link: https://www.dof.gob.mx/nota_detalle.php?codigo=5653931&fecha=02/06/2022#gsc.tab=0

```python
p = [0.2, 0.3, 0.5]
n = [3, 5, 10]
promedio_retornos = calculate_weighted_returns(SB, p, n)

# Encuentra la fila con el valor más alto en 'Weighted Average Return'
fila_superior = promedio_retornos.sort_values(by='Weighted Average Return', ascending=False).iloc[0]

# Extrae el nombre y el valor
nombre = fila_superior['Siefore']
valor_alto = fila_superior['Weighted Average Return']

print(f"El siefore: {nombre}, tiene el rendimiento más alto: {valor_alto}")
promedio_retornos
```
